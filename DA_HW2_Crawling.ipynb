```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By 
from webdriver_manager.chrome import ChromeDriverManager
import time
import openpyxl
```

```python
#sheet 생성
wb = openpyxl.Workbook()
sheet = wb.active
sheet.append(["제목", "아이디", "리뷰 내용"])
```

```python
service = Service(executable_path=ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)
```

```python
# yes24 이동
url = 'http://www.yes24.com/main/default.aspx'
browser.get(url)
time.sleep(1)
```

```python
# 헤어질 결심 검색
search = browser.find_element(By.XPATH, "//*[@id='query']")

search.send_keys('헤어질 결심')
time.sleep(1)
search.send_keys(Keys.ENTER)
time.sleep(1)

# 헤어질 결심 페이지 이동
browser.find_element(By.XPATH, '//*[@id="yesSchList"]/li[1]/div/div[2]/div[2]/a[1]').click()
```

```python
# 첫 번째 리뷰 페이지 크롤링
for j in range(5):
    title = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > span.review_tit > span'.format(j+3)).text
    nickname = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > div > em.txt_id > a'.format(j+3)).text
    review = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoBot.crop > a > div'.format(j+3)).text
    sheet.append([title,nickname,review])
```

```python
for i in range(4,13):
    browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > a:nth-child({})'.format(i)).click()

    time.sleep(1)
    for j in range(5):
        title = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > span.review_tit > span'.format(j+3)).text
        nickname = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > div > em.txt_id > a'.format(j+3)).text
        review = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoBot.crop > a > div'.format(j+3)).text
        sheet.append([title,nickname,review])
```

```python
# 다음 페이지 버튼 클릭
browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > a.bgYUI.next').click()
```

```python
#마지막 장 크롤링
for j in range(5):
    title = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > span.review_tit > span'.format(j+3)).text
    nickname = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > div > em.txt_id > a'.format(j+3)).text
    review = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoBot.crop > a > div'.format(j+3)).text
    sheet.append([title,nickname,review])

browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > a:nth-child(4)').click()
time.sleep(1)
for j in range(5):
    title = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > span.review_tit > span'.format(j+3)).text
    nickname = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > div > em.txt_id > a'.format(j+3)).text
    review = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoBot.crop > a > div'.format(j+3)).text
    sheet.append([title,nickname,review])

total_reviews = int(browser.find_element(By.XPATH, '//*[@id="emReviewCountText"]').text[1:3])

# 마지막 페이지 리뷰 수
last_page_reviews = total_reviews%5

browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div.review_sort.sortTop > div.review_sortLft > div > a:nth-child(5)').click()
time.sleep(1)    
for j in range(last_page_reviews):
    title = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > span.review_tit > span'.format(j+3)).text
    nickname = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoTop > div > em.txt_id > a'.format(j+3)).text
    review = browser.find_element(By.CSS_SELECTOR, '#infoset_reviewContentList > div:nth-child({}) > div.reviewInfoBot.crop > a > div'.format(j+3)).text
    sheet.append([title,nickname,review])
```

```python
browser.close()
```

```python
for row in sheet.iter_rows(max_col=3, values_only=True):
    print(row)
```
