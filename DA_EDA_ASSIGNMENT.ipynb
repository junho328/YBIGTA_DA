```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = pd.read_csv('/content/wine.csv')
wine.head()
```

# Wine Data

- 각 요인이 와인의 품질에 어떻게 영향을 미치는지 분석하고자 함.
- 여러 요인 중에서도 pH,fixed acidity,그리고 citric acid에 대해서 분석함.

```python
# pH, fixied acidity, citric acid 열만 추출
wine = wine[['quality','pH','fixed acidity','citric acid','type']]
```

## 결측치 제거

```python
# 결측치 확인
wine[wine.isnull()]
```

```python
# 결측이 있는 행 제거
wine.dropna()
```

```python
#레드와인과 화이트 와인을 구분
red_wine = wine[wine['type']=='red']
white_wine = wine[wine['type']=='white']
```

```python
# type 제거
red_wine.drop(['type'],axis=1,inplace=True)
white_wine.drop(['type'],axis=1, inplace=True)
```

## 시각화1 - Pair Plot

```python
# red wine의 pairplot
sns.pairplot(red_wine)
plt.title("Red Wine Data의 Pair Plot")
plt.show()
```

```python
# white wine의 pairplot
sns.pairplot(white_wine)
plt.title("White Wine Data의 Pair Plot")
plt.show()
```

## 2. 시각화2 - Scatter Plot

```python
# red wine의 scatter plots
```

```python
plt.figure(figsize=(10,10))
plt.subplot(311)
plt.scatter(red_wine['pH'],red_wine['quality'])
plt.title('pH - quality')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(312)
plt.scatter(red_wine['fixed acidity'],red_wine['quality'])
plt.title('fixed acidity - quality')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(313)
plt.scatter(red_wine['citric acid'],red_wine['quality'])
plt.title('citric acid - quality')
plt.show()
```

```python
# white wine의 scatter plots
```

```python
plt.figure(figsize=(10,10))
plt.subplot(311)
plt.scatter(white_wine['pH'],white_wine['quality'])
plt.title('pH - quality')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(312)
plt.scatter(white_wine['fixed acidity'],white_wine['quality'])
plt.title('fixed acidity - quality')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(313)
plt.scatter(white_wine['citric acid'],white_wine['quality'])
plt.title('citric acid - quality')
plt.show()
```

# 시각화 3. Heatmap을 이용한 상관관계 분석

```python
# red wine 상관관계 분석
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data = red_wine.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='OrRd')
```

```python
# white wine 상관관계 분석
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data = white_wine.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='Greys')
```

# 결론
- pH,fixed acidity, citric acid 각각의 요인과 quality(품질) 사이에 상관관계가 작았다.
- 하지만 각 요소에 따른 scatter plot을 그려본 결과, 어느 정도 유의미한 관계가 있음을 확인할 수 있었다.
  - red wine : pH가 3.2 - 3.6 사이일 때 높은 품질을 보인다
  - white wine : pH가 3.0 - 3.4 사이일 때 높은 품질을 보이며, red wine과 다르게 citric acid,fixed acidity와 품질 간 관계가 더 잘 드러난다.
